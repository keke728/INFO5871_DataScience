```python
import pandas as pd
import numpy as np 
import matplotlib
```

### Import Datasets

Make dataframes for both csv files and drop the duplicates in demographics by user_id .

```python
response_df = pd.read_csv('response.csv')
demographics_df = pd.read_csv('demographics.csv')
demographics_df = demographics_df.drop_duplicates(subset='user_id',keep='last')
response_df.shape
```

### Data Cleaning

##### 1. Merge dataframes on "user_id".

```python
cc_df = pd.merge(response_df, demographics_df, on='user_id')
print(cc_df.shape)
```

##### 2. Check on the data with the "groupby" method on 'user_id',  there should be 39 trials for each participant.

```python
cc_df.groupby('user_id').size()
```

##### 3. Something went wrong for the subject with id 47, who seemed to have done this study twice, drop their second result.

```python
user47 = cc_df[cc_df['user_id']==47]
user47 = user47.drop_duplicates(subset='trial_number',keep='first')
```

```python
df = cc_df[cc_df["user_id"] != 47]
df.shape
```

```python
cc_df = pd.concat([df, user47], ignore_index=True)
print(cc_df.shape)
cc_df.head()
```

##### 4. Check on the normal_vision column, drop "no" if there is any.

```python
cc_df[cc_df['normal_vision']=='no']
```

```python
cc_df = cc_df[cc_df.user_id != 12]
```

##### 5. Drop entries for engagement check

```python
cc_df = cc_df[cc_df.model_type != 'engagement']
```

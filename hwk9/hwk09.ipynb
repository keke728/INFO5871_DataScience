## Connect to evals database

Execute a query extracting score, ethnicity, gender, language, age, cls_perc_eval, cls_students, cls_level, bty_avg, and pic_outfit from the evals table dropping the one credit classes, the teaching faculty and the classes with more than 400 students.

```python
eng = sqla.create_engine('sqlite:///evals.db')
```

```python
query = 'SELECT score, \
                ethnicity, \
                gender, \
                language, \
                age, \
                cls_perc_eval, \
                cls_students, \
                cls_level, \
                bty_avg, \
                pic_outfit \
          FROM evals \
          WHERE rank = "tenured" OR rank = "tenure track" AND \
                cls_credits = "multi credit" AND \
                cls_students < 400 OR cls_students = 400'

evalsdf = pd.read_sql_query(query, eng)
evalsdf.head()
```

## Do a pairplot to explore the data

```python
sns.pairplot(evalsdf);
```

## Multiple regression

Using backwards-selection by p-value, produce a minimal model

```python
model1 = sm.ols(formula ="score ~ ethnicity + gender + language + age + cls_perc_eval + cls_students + cls_level + bty_avg + pic_outfit", data=evalsdf)
reg1 = model1.fit()
print(reg1.aic)
reg1.summary()
```

```python
model2 = sm.ols(formula ="score ~ ethnicity + gender + language + age + cls_perc_eval + cls_level + bty_avg + pic_outfit", data=evalsdf)
reg2 = model2.fit()
reg2.aic
```

```python
model3 = sm.ols(formula ="score ~ ethnicity + gender + language + age + cls_perc_eval + bty_avg + pic_outfit", data=evalsdf)
reg3 = model3.fit()
reg3.aic
```

```python
model4 = sm.ols(formula ="score ~ ethnicity + gender  + age + cls_perc_eval + bty_avg + pic_outfit", data=evalsdf)
reg4 = model4.fit()
reg4.aic
```

```python
model5 = sm.ols(formula ="score ~ ethnicity + gender  + age + cls_perc_eval + bty_avg", data=evalsdf)
reg5 = model5.fit()
reg5.aic
```

## Question 4: Assessing seasonal variation

Using the $\chi^2$ method, determine if this observed seasonal variation is statistically significant.

### What is the null hypothesis?

_H0: The observed seasonal variation is a random sample, there is no significant seasonal variation between the observed value and the expected value, and the observed variation reflects natural sampling fluctuation._

### Call the `chisquare` function to do the calculation

From the result, print the $\chi^2$ statistic. `res.statistic` and the p-value `res.pvalue`.

```python
total_flights_2006 = total_flights_2006.reset_index(drop=True)
expected = [47253.5, 47253.5, 47253.5, 47253.5, 47253.5, 47253.5, 47253.5, 47253.5, 47253.5, 47253.5, 47253.5, 47253.5]
observed = total_flights_2006['Total Flights'].tolist()
```

```python
st.chisquare(observed, expected)
```

### Conclusion

_The $\chi^2$ statistic calculated value is way bigger than the $\chi^2$ critical value, and p=0, so we reject the null hypothesis, and conclude that
there is a significant seasonal variation between the observed value and the expected value._

## Question 5: Has the seasonal variation changed from 2006 to 2018?

### What is the null hypothesis?

_The seasonal variation remained the same and didn't change from 2006 to 2018._

### Extract the same data for 2018

May is in there twice. Drop the duplicate

```python
month_filter18 = (flt_df.index > '2017-12-01') & (flt_df.index < '2019-01-01')
total_flights_2018 = flt_df.loc[month_filter18].copy()
total_flights_2018.drop_duplicates(keep = 'last', inplace = True)
total_flights_2018
```

### Combine with above in a data frame

Use `reset_index(drop=True)` first. This will remove the index and you'll just have integer indices.

```python
seasonal_df = total_flights_2006.append(total_flights_2018, ignore_index=True)
seasonal_df.drop(seasonal_df.columns[0:7], axis=1, inplace=True)
seasonal_df
```

### Create a projected 2006 column

This is the number of flights for each month if the distribution was the same as 2006. For example, January projected = January 2006 / Total for 2006 * Total of 2018. That way the count for 2018 and the projected count are the same, but the project column has the proportional distribution of 2006.

We need to do this because the $\chi^2$ statistic depends on the counts and if the totals are different, the results aren't meaningful.

```python
seasonal_df['Projected'] = (total_flights_2006['Total Flights'] / total_flights_2006['Total Flights'].sum())*total_flights_2018['Total Flights'].sum()
seasonal_df
```

### Create a side-by-side bar plot

See attached PNG file.

```python
data = {'2006':seasonal_df['Projected'].loc[:11], '2018':seasonal_df['Total Flights'].loc[12:]}

# set height of bar
bars1 = data['2006'].tolist()
bars2 = data['2018'].tolist()

# set width of bar
barWidth = 0.3

# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

# Make the plot
plt.bar(r1, bars1,  width=barWidth, label='2006')
plt.bar(r2, bars2,  width=barWidth, label='2018')

# Create legend & Show graphic
plt.legend(bbox_to_anchor=(1.3, 0.8),loc='right')
plt.show()
```

### Conclusion

_The seasonal variation remained the same and didn't change from 2006 to 2018, we didn't reject the null hypothesis._
